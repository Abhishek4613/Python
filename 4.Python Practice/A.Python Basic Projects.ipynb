## Email ID Validation

```python
def EmailValidation():
    
    z = input("Enter Email ID: ").lower()
    
    c = 0
    
    for i in z:
        
        if i == "@":
            c+= 1
            
    if c == 1:

        if z[-4:] == ".com":
            
            if z[0].isalpha():

                x = z.index("@")

                if z[:x].isalnum() or z[:x] in ".":

                    if z[x+1:-4].isalpha():
                        
                        if len(z[:x])>=6 and len(z[:x])<=12:

                            print("Your Email ID created successfully. \nYour Email ID: ",z)

                        else:
                            print("Invalid User Name!!! User name must be between 6 to 12 characters.")

                    else:
                        print("Invalid Domain Name!!! Check domain name.")

                else:
                    print("Invalid User Name!!! User name should be combination of alphabates and numbers and \".\".")
    
            else:
                print("Invalid User Name!!! User name should always start with alphabates.") 
    
        else:      
            print("Invalid Domain Name!!! Check domain name.")
        
    else:
        print("Invalid Email ID!!! There must be one and only one \"@\".")
```

## Create Valid Password

```python
def CreatePassword():
    
    while True: 
        
        password1 = input("Create Your Password: ")
        password2 = input("Enter Your Password: ")
    
        if password1 == password2:
            
            if len(password1) >= 8 and len(password1) <=16:
                
                
                count_upper = 0
                count_lower = 0
                count_number = 0
                count_spec_char = 0
                
                
                for i in password1:
        
                            if i.isupper():
                                count_upper+=1
            
                            elif i.islower():
                                count_lower+=1
            
                            elif i.isnumeric():
                                count_number+=1
                
                            else:
                                count_spec_char+=1
                            
                if count_upper >=1:
                    
                    if count_lower >=1:
                        
                        if count_number >=1:
                            
                            if count_spec_char >=1:
                                
                                print("Your Password Created Successfully.")
                                
                                break
                                
                            else:
                                print("Invalid Password, Password Must Contain a Special Charecter.")
                            
                        else:
                            print("Invalid Password, Password Must Contain a Number.")
                    
                    else:
                        print("Invalid Password, Password Must Contain a Lower Cased Later.")   
                
                else:
                    print("Invalid Password, Password Must Contain a Upper Cased Later.")      
            
            else:
                
                print("Invalid Password, password length must be in between 8 to 16.")     
        else:
            print("Password Mismatched. Enter Again.")  
```

## ToDo List

```python
tasks = []


def add_task(task):
    tasks.append(task)
    
    
def remove_task(index):
    del tasks[index]
    

def show_tasks():
    print("Your Current Tasks:")
    
    for index,task in enumerate(tasks):
        print(f"{index+1}.{task}")
        
def main():
    
    while True:
        
        print("What you would like to do ?")
        print("1. Add a Task")
        print("2. Remove a Task")
        print("3. View All Task")
        print("4. Quit")
        
        choice = int(input("Enter Your Choice: 1, 2, 3, or 4: "))
        
        if choice == 1:
            
            task = input("Enter Task Name: ")
            
            add_task(task)
            
            print("Task Added")
            
            
        elif choice == 2:
            
            task_index = int(input("Enter Task Number of Task to Remove: "))-1
            
            remove_task(task_index)
            
            print("Task Removed")
          
         
        elif choice == 3:
            
            show_tasks()
            
            
        elif choice == 4:
            
            print("Good Bye, Have a Nice Day")
            
            break
            
               
        else:
            print("Invalid Entry, Try Again")
```

## Guess The Movie Name

```python
import random

def GuessTheMovieName():
    
    movies = ["The Kashmir Files","Brahmastra","Vikram Vedha","Laal Singh Chaddha","Ram Setu","Gehraiyaan","Gangubai Kathiawadi","Phone Bhoot","Drishyam","Ram Setu","Goodbye","Cuttputlli","Darlings","Jersey"]

    movie = random.choice(movies)
    
    movie = movie.replace(" ","").lower()

    print("Hello, Welcome to the Game: \"Guess The Movie Name\"")
    print("You have to Guess the Name of Bollywood Movie with the hlep of Hint.")
    


    hint = "_"*len(movie)


    for i in range(len(movie)):
    
        if movie[i] in "aeiou":
        
            hint = hint[:i] + movie[i] + hint[i+1:]
        
    print("\nHint: ",hint, ("Word Length of Movie Name: ",len(movie)))

    print("\n\"You have to enter all charecters in once. Do not leave any space betwee them.\"")
    
    attempts = 0
    
    while attempts < 5 and hint != movie:
        
        guess = input("\nEnter the Characters: ").lower()
        
        if len(guess) != len(movie):
            print("Invalid Input. Please Enter", len(movie)-(len(hint)-hint.count("_")), "Characters.")
            continue
        
        correct = True
        
        for i in range(len(guess)):
            if guess[i] != movie[i]:
                correct = False
                break
        
        if correct:
            hint = movie
            print("\nWow!!! You Won")
        else:
            print("\nSorry,You Lost. Try Again")
            attempts += 1
            
    if attempts == 5:
        print("\nYou have used all 5 attempts. The movie was", movie)
```
